```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.worlddata.info/largest-economies.php'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html.parser')                    
```

```python
print(soup)
```

```python
soup.find_all("table")
```

```python
#COUNTRIES NAMES
soup.find_all("td")
n = soup.find_all('a')[37:87]
names = [i.text for i in n]
names
```

```python
#Rank
soup.find_all('tr')[1:]
m=soup.find_all("td")[0:][0:][0::1][0::2][0::2]
rank= [i.text for i in m]
rank
```

```python
#GDP
m=soup.find_all("td")[0:][1:][1::2][0::2]
gdp= [i.text for i in m]
gdp
```

```python
# Per Capital
soup.find_all('tr')[1:]
n= soup.find_all('td')[1::2][1::2]

cap = [i.text for i in n]
cap
```

```python
import pandas as pd
```

```python
Data = {'Rank':rank,'Country':names,'GDP_Billions$':gdp,'Per_Capital$':cap} 
```

```python
df=pd.DataFrame(Data)
df
```
