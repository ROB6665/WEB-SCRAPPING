Importing libraries

```python
import pandas as pd
import selenium as se
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
```

Get the website

```python
website = "https://play.google.com/store/apps/?hl=en_IN&gl=US&pli=1"
```

Opening Website

```python
driver = webdriver.Chrome()
driver.get(website) 
```

Click on search button

```python
search = driver.find_element(By.CLASS_NAME,'oc9n0c')
Search_button = search.find_element(By.TAG_NAME,'button')
Search_button.click()
```

Input the app name

```python
app = input('App Name = ')
input = driver.find_element(By.TAG_NAME,'input')
input.send_keys(app)
```

Select the first app

```python
results = driver.find_element(By.CLASS_NAME,'XPmdRc')
results.click()
```

Open the app

```python
click_on_app = driver.find_element(By.CLASS_NAME,'ipRz4')
click_on_app.click()
```

See all review

```python
reviews = driver.find_element(By.CLASS_NAME,'Jwxk6d')
reviews.find_elements(By.TAG_NAME,'div')
button = reviews.find_element(By.CLASS_NAME,'u4ICaf')
click_button = button.find_element(By.TAG_NAME,'span')
```

```python
click_button.click()
```

Select the data 

```python
driver.find_element(By.CLASS_NAME,'jgIq1')
```

Extract reviews data

```python
driver.find_elements(By.CLASS_NAME,'RHo1pe')
```

Names of the customer

```python
Names = driver.find_elements(By.CLASS_NAME,'X5PpBb')
names= []
for i in Names:
    names.append(i.text)
```

```python
Names = names[2:]
Names
```

Reviews of the customer

```python
Data = driver.find_elements(By.CLASS_NAME,'h3YV2d')
Reviews= []
for i in Data:
    Reviews.append(i.text)
```

```python
Reviews = Reviews[2:]
Reviews
```

Date of reviews

```python

Data = driver.find_elements(By.CLASS_NAME,'Jx4nYe')
Date= []
for i in Data:
    Date.append(i.text)
```

```python
Date = Date[2:]
Date
```

Creating Dict

```python
DT = {'Name':Names,'Date':Date,'Reviews':Reviews}
```

```python
DT
```

Creating Data frame

```python
df = pd.DataFrame(DT)
```
