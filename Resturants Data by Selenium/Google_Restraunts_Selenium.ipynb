Opening Browser

```python
driver = webdriver.Chrome()
driver.get(website)
```

All Names

```python
Restraunts_Names = driver.find_elements(By.CLASS_NAME,'OSrXXb')[3:]
Names=[]
for data in Restraunts_Names:
    names = data.text
    Names.append(names)
Names=Names[0:20]
Names
```

After Selecting Restraunt from the list. 
Extracted From right side ( Search Part ) [ Name and contact ]

```python
Data = driver.find_elements(By.CLASS_NAME,'AVvGRc')
for data in Data:
    DT = data.text.split('\n')
Name = DT[0]
Contact = DT[18] # In case number is not showing try choose [17,18,19,20,21,22]
```

```python
Name_Contact  =  {'Restraunt': Name,'Contact': Contact}
Name_Contact
```

All Details from left side list

```python
Left_basic = driver.find_elements(By.CLASS_NAME,'rllt__details')
l=[]
for data in Left_basic:
    Restraunts = data.text.split('\n')
    l.append(Restraunts)
l
```

Creating DataFrame and Save it as CSV File

```python
df=pd.DataFrame(l)
df
# df.to_csv('Resturants_data.csv',index=False)
```
